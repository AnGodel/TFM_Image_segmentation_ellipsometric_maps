```python
def compareSegmentedmaps(Dmap, Pmap):
    cluster_pairs = []
    for Dc_label in np.unique(firstD):
        Dcoordinates = np.where(firstD==Dc_label)
        for Pc_label in np.unique(firstP):
            PCoordinates = np.where(firstP==Pc_label)
            if PCoordinates == Dcoordinates:
                D_Ptuple = (Dc_label, Pc_label)
                cluster_pairs.append(D_Ptuple)
    return cluster_pairs
```
