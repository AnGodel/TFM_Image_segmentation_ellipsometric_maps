```python
from skimage.color import rgb2gray
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
from scipy import ndimage
```

```python
image = plt.imread('./1.jpeg')
image.shape
```

```python
image[191,262]
```

```python
plt.imshow(image);
```

```python
gray = rgb2gray(image)
plt.imshow(gray, cmap='gray')
```

```python
gray.shape
```

```python
gray_r = gray.reshape(gray.shape[0]*gray.shape[1])
for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 1
    else:
        gray_r[i] = 0
gray = gray_r.reshape(gray.shape[0],gray.shape[1])
plt.imshow(gray, cmap='gray')
```

```python
gray_r = gray.reshape(gray.shape[0]*gray.shape[1])
for i in range(gray_r.shape[0]):
    if gray_r[i] > gray_r.mean():
        gray_r[i] = 3
    elif gray_r[i] > 0.5:
        gray_r[i] = 2
    elif gray_r[i] > 0.25:
        gray_r[i] = 1
    else:
        gray_r[i] = 0
gray = gray_r.reshape(gray.shape[0],gray.shape[1])
plt.imshow(gray, cmap='gray')
```

## Convolution Explained

```python
image = plt.imread('index.png')
```

```python
plt.imshow(image)
```

```python
gray = rgb2gray(image)

# defining the sobel filters
sobel_horizontal = np.array([np.array([1, 2, 1]), np.array([0, 0, 0]), np.array([-1, -2, -1])])
print(sobel_horizontal, 'is a kernel for detecting horizontal edges')
 
sobel_vertical = np.array([np.array([-1, 0, 1]), np.array([-2, 0, 2]), np.array([-1, 0, 1])])
print(sobel_vertical, 'is a kernel for detecting vertical edges')
```

```python
out_h = ndimage.convolve(gray, sobel_horizontal, mode='reflect')
out_v = ndimage.convolve(gray, sobel_vertical, mode='reflect')
# here mode determines how the input array is extended when the filter overlaps a border.
```

```python
plt.imshow(out_h, cmap='gray')
```

```python
plt.imshow(out_v, cmap='gray')
```

```python
kernel_laplace = np.array([np.array([1,1,1]), np.array([1,-8,1]), np.array([1,1,1])])
print(kernel_laplace, 'is a laplacian kernel')
```

```python
out_1 = ndimage.convolve(gray, kernel_laplace, mode='reflect')
plt.imshow(out_1, cmap='gray')
```

```python
pic = plt.imread('1.jpeg')/255
print(pic.shape)
```

```python
pic[191,262]
```

```python
pic_n = pic.reshape(pic.shape[0]*pic.shape[1], pic.shape[2])
pic_n.shape
```

```python
pic_n[191]
```

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=5, random_state=0).fit(pic_n)
pic2show = kmeans.cluster_centers_[kmeans.labels_]
```

```python
cluster_pic = pic2show.reshape(pic.shape[0], pic.shape[1], pic.shape[2])
plt.imshow(cluster_pic)
```

```python
kmeans2 = KMeans(n_clusters=10, random_state=0).fit(pic_n)
pic2show2 = kmeans2.cluster_centers_[kmeans2.labels_]
cluster_pic2 = pic2show2.reshape(pic.shape[0], pic.shape[1], pic.shape[2])
plt.imshow(cluster_pic2)
```

Se ve que el nmero de clusters elegido para el kmeans influye mucho en la segmentacin. No le veo utilidad directa aplicada a mis imgenes porque habra que ir probando manualmente con distintas cantidades de clsters.
