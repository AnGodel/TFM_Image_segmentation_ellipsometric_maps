```python
C_Selector = 2

idx = xxx.n_clustersList_shuffled[C_Selector]
Deltas = xxx.delta_shot
Psis = xxx.psi_shot
WL = xxx.WLarray

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,5))
fig.tight_layout()

ax1.clear
ax1.scatter(xxx.WLarray, Deltas[C_Selector], color='red')
ax1.set_title('Delta')
fig.suptitle('Cluster index: {}'.format(C_Selector), y=1.05)    
ax2.scatter(xxx.WLarray, Psis[C_Selector], color='blue')
ax2.set_title('Psi')

for C_Selector in xxx.n_clustersList_shuffled:
    ax1.plot(xxx.WLarray, Deltas[C_Selector], alpha=0.7)
    ax2.plot(xxx.WLarray, Psis[C_Selector], alpha=0.7)
```

```python
idxSelector = 0
C_Selector = 2

Dmap = xxx.segmentedShuffledStack[:,:,xxx.DeltaIndices[idxSelector]]
Pmap = xxx.segmentedShuffledStack[:,:,xxx.PsiIndices[idxSelector]]

D_Cvalues, D_Ccounts = np.unique(Dmap, return_counts = True)
P_Cvalues, P_Ccounts = np.unique(Pmap, return_counts = True)
D_varwidth = (D_Cvalues.max() - D_Cvalues.min())/20
P_varwidth = (P_Cvalues.max() - P_Cvalues.min())/20
SelectedDval = xxx.delta_shot[C_Selector][xxx.WLIndices[idxSelector]]
SelectedPVal = xxx.psi_shot[C_Selector][xxx.WLIndices[idxSelector]]

fig, (ax1, ax2) = plt.subplots(1,2, figsize=(15,5))
fig.tight_layout()

ax1.clear
ax1.bar(D_Cvalues, D_Ccounts, 
        edgecolor='green',
        linewidth=1.5, 
        width=D_varwidth,
        color='red')
ax1.axvline(x=SelectedDval, color='black', label='cluster {} at {}'.format(C_Selector, SelectedDval))
ax1.legend()
ax1.set_title('Delta')
fig.suptitle('Map index: {}: Wavelength: {} nm'.format(idxSelector, xxx.WLdict[idxSelector]), y = 1.05)
ax2.bar(P_Cvalues, P_Ccounts, 
        edgecolor='green', 
        linewidth=1.5, 
        width=P_varwidth,
        color='blue')
ax2.axvline(x=SelectedPVal, color='black', label='cluster {} at {}'.format(C_Selector, SelectedPVal))
ax2.legend()
ax2.set_title('Psi')
```

```python
idxSelector = 2
C_Selector = 2

Dmap = xxx.AllShuffledStack[:,:,xxx.DeltaIndices[idxSelector]]
Pmap = xxx.AllShuffledStack[:,:,xxx.PsiIndices[idxSelector]]
DSegmap = xxx.segmentedShuffledStack[:,:,xxx.DeltaIndices[idxSelector]]
PSegmap = xxx.segmentedShuffledStack[:,:,xxx.PsiIndices[idxSelector]]

C_ys, C_xs = xxx.cluster_coordinates[C_Selector]

fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(nrows=2,ncols=2, figsize=(12,16))
fig.tight_layout(pad=3)
fig.suptitle('Cluster {} at Wavelength {}, {} nm'.format(C_Selector, 
                                                         idxSelector,
                                                         xxx.WLdict[idxSelector]),
            y = 1.02)
plt.ylim(ymin=xxx.dim1all, ymax=0)
plt.xlim(xmin=0, xmax=xxx.dim2all)
ax1.clear
ax1.imshow(Dmap, cmap='gray')
ax1.set_title('Raw Delta map')
ax1.scatter(C_xs, C_ys, s=5, color='pink')
ax1.grid(False)

ax2.clear
ax2.imshow(Pmap, cmap='gray')
ax2.set_title('Raw Psi map')
ax2.scatter(C_xs, C_ys, s=5, color='pink')
ax2.grid(False)

ax3.clear
ax3.grid(False)
ax3.imshow(DSegmap, cmap='viridis')
ax3.set_title('Segmented Delta map')
ax3.scatter(C_xs, C_ys, s=5, color='pink')

ax4.clear
ax4.imshow(PSegmap, cmap='viridis')
ax4.set_title('Segmented Psi map')
ax4.scatter(C_xs, C_ys, s=5, color='pink')
ax4.grid(False)
```
